```python
#Plot of CDF

counts, bin_edges = np.histogram(irish_setosa['petal_length'], bins=10, density = True)
pdf= counts/(sum(counts))
print(pdf);
print(bin_edges);
cdf= np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf);
plt.plot(bin_edges[1:],cdf)
counts, bin_edges = np.histogram(irish_setosa['petal_length'], bins=20, density= True)
pdf = counts/(sum(counts))
plt.plot(bin_edges[1:],pdf);

plt.show();
```

```python
#What is PDF?
#Probabilty Density Function
#Kernel density estimation - KDE
#Percentage of total data point ar the value is PDF
#CDF- cumulative density function
#CDF at x0 is defined as percentage of all data point below x0
#Therefore Intrgration of PDF= CDF
```

```python
print("Mean :")
print (np.mean(irish_setosa["petal_length"]))

print(np.mean(np.append(irish_setosa["petal_length"],50)))
print (np.mean(irish_virginica["petal_length"]))
print (np.mean(irish_versicolor["petal_length"]))

print('\nStd-dev :');
print(np.std(irish_setosa["petal_length"]));
print(np.std(irish_virginica["petal_length"]));
print(np.std(irish_versicolor["petal_length"]));
```

```python
print ("median :")
print (np.median(irish_setosa["petal_length"]))

print(np.median(np.append(irish_setosa["petal_length"],50)));
print(np.median(irish_virginica["petal_length"]))
print(np.median(irish_versicolor["petal_length"]))

print("\nQuantiles: ")
print(np.percentile(irish_setosa["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(irish_virginica["petal_length"], np.arange(0, 100, 25)))
print(np.percentile(irish_versicolor["petal_length"], np.arange(0, 100, 25)))

print("\n 90th Percentitles:")
print(np.percentile(irish_setosa["petal_length"],90))
print(np.percentile(irish_virginica["petal_length"],90))
print(np.percentile(irish_versicolor["petal_length"],90))

from statsmodels import robust
print("\n Median Absolute Deviation")
print(robust.mad(irish_setosa["petal_length"]))
print(robust.mad(irish_virginica["petal_length"]))
print(robust.mad(irish_versicolor["petal_length"]))
```
